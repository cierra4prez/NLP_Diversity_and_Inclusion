# Scraper suppliment
After scraping several thousand urls from glassdoor.com's main [Browse Companies](https://www.glassdoor.com/Explore/browse-companies.htm?overall_rating_low=0&page=1&isHiringSurge=0) page, the data base which was created was primarily comprised of companies which had over 10,000 employees. Glassdoor will not allow the scraper to surpass page 389 using the root url, and instead resets to page 1. 

In order to work around Glassdoor's block and address the company size imbalance, this `scraper_suppliment` notebook was created to customize the "Global Company Size" filter.

**Process:** first scrape webpage urls, then run those urls through the scraper function
1. From [main url](https://www.glassdoor.com/Explore/browse-companies.htm?overall_rating_low=0&page=1&isHiringSurge=0), manually choose "Company Size 501 to 1000 Employees."
2. Scrape urls on this page.
    - driver to click "Next" button, then scrape urls and repeat until desired number of urls have been collected

How many urls to scrape in order to balance data frame?

- 800 urls for 501-1k
- 800 urls for 201-500
- 400 urls for 51-200
- 400 urls for 1-50
- 400 urls for 5k-10k

*Note: each of the above groups are gathered separately, and so below code is repetetive. In order to reproduce results, inspect code for one of these groups.*

**Imports**

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

import pandas as pd
import numpy as np
import time
```

**Setting up User-Agent to prevent ip block:**

```python
opts = Options()
opts.add_argument("Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.96 Safari/537.36")
#opts.add_argument('headless')

driver = webdriver.Chrome(options=opts)
time.sleep(5)
```

**DRIVER HERE:**

```python
#start driver on main url, 
url_main = 'https://www.glassdoor.com/Explore/browse-companies.htm?overall_rating_low=0&page=1&isHiringSurge=0' #main url

driver.get(url_main)

## manually click on company size option
```

```python
# uncomment this cell to empty the list
#urls_scraped_501to1k = []
```

```python
while len(urls_scraped_501to1k) <800: #### UPDATE VALUE #### UPDATE VALUE #### UPDATE VALUE #### UPDATE VALUE #### UPDATE VALUE ##
    #scrape page for urls
    elems = driver.find_elements_by_tag_name('a') #find links on an individual search page tagged with the 'a' tag

    #list of all links on page
    company_links = [] #all links
    for elem in elems:
        company_link = elem.get_attribute('href') #returns every item with 'href' attribute (these are the links for each company)
        company_links.append(company_link)
        #iterating through each company's "Overview" url
    for url in company_links:
        if 'Overview' in url:
            urls_scraped_501to1k.append(url) #### UPDATE VALUE #### UPDATE VALUE #### UPDATE VALUE #### UPDATE VALUE #### UPDATE VALUE ##

    #click "Next"
    next_button = driver.find_element_by_xpath('//*[@id="ReactCompanyExplorePageContainer"]/div/div/div/div/div[3]/div/ul/li[7]/button')
    next_button.click()
    time.sleep(3)
```
